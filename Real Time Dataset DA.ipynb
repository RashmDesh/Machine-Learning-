# Linear Regression



on trains,csv dataset

x=independant variable=Dept,Store

y=dependant variable=Weekly_Sales

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train.isnull().any()
```

```python
#droping data that we do not want to find corr
data=train.drop(['Weekly_Sales','IsHoliday','Dept','Date'],axis=1)
```

```python
x_train,x_test, y_train, y_test=train_test_split(data,train['Weekly_Sales'],test_size=0.20,random_state=8)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
accuracy=model.score(x_test,y_test)
print(accuracy*100,'%')
```

```python
#Prediction of Test and Training set result  
y_pred= model.predict(x_test)  
x_pred= model.predict(x_train)  
```

```python
y_pred
```

```python
x_pred
```

```python
#visualizing the Training set results
plt.scatter(x_train, y_train, color="green")   
plt.plot(x_train, x_pred, color="red")    
plt.title("Train (Training Dataset)")  
plt.xlabel("Store")  
plt.ylabel("weekly sales")  
plt.show() 
```

```python
##visualizing the Test set results 
plt.scatter(x_test, y_test, color="blue")   
plt.plot(x_train, x_pred, color="red")    
plt.title("Weekly sales vs Store (Test data)")  
plt.xlabel("Store")  
plt.ylabel("weekly sales")  
plt.show()  
```
